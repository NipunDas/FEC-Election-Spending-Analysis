```python
import polars as pl
```

```python
# source: https://www.fec.gov/campaign-finance-data/pac-and-party-summary-file-description/
pac_col_names = ['cmte_id', 'cmte_nm', 'cmte_tp', 'cmte_dsgn', 'cmte_filing_freq', 'ttl_receipts', 'trans_from_aff', 'indv_contrib', 'other_pol_cmte_contrib', 'cand_contrib', 'cand_loans', 'ttl_loans_received', 'ttl_disb', 'tranf_to_aff', 'indv_refunds', 'other_pol_cmte_refunds', 'cand_loan_repay', 'loan_repay', 'coh_bop', 'coh_cop', 'debts_owed_by', 'nonfed_trans_received', 'contrib_to_other_cmte', 'ind_exp', 'pty_coord_exp', 'nonfed_share_exp', 'cvg_end_dt']

pac_df = pl.read_csv('./data/webk26.txt', separator='|', has_header=False, new_columns=pac_col_names, ignore_errors=True)
```

```python
# View PAC data frame
pac_df
```

```python
# Check for missing values
pac_df.null_count()
```

```python
# source: https://www.fec.gov/campaign-finance-data/contributions-committees-candidates-file-description/
contributions_col_names = ['cmte_id', 'amndt_ind', 'rpt_tp', 'transaction_pgi', 'image_num', 'transaction_tp', 'entity_tp', 'name', 'city', 'state', 'zip_code', 'employer', 'occupation', 'transaction_dt', 'transaction_amt', 'other_id', 'cand_id', 'tran_id', 'file_num', 'memo_cd', 'memo_text', 'sub_id']
contributions_df = pl.read_csv('./data/itpas2.txt', separator='|', has_header=False, new_columns=contributions_col_names, ignore_errors=True)
```

```python
# View contributions dataframe
contributions_df
```

```python
# Check for missing values
contributions_df.null_count()
```

```python
# View transaction types
contributions_df['transaction_tp'].value_counts()
```

```python
# Use FEC transaction types to convert codes to more meaningful descriptions
transaction_type_map = {
  '24A': 'Independent expenditure opposing election of candidate',
  '24C': 'Coordinated party expenditure',
  '24E': 'Independent expenditure advocating expenditure of candidate',
  '24F': 'Communication cost for candidate',
  '24K': 'Contribution made to nonaffiliated committee',
  '24Z': 'In-kind contribution made to registered filer',
}

transaction_df = pl.DataFrame({
  'transaction_tp': ['24A', '24C', '24E', '24F', '24K', '24Z'],
  'transaction_description': [
    'Independent expenditure opposing election of candidate',
    'Coordinated party expenditure',
    'Independent expenditure advocating expenditure of candidate',
    'Communication cost for candidate',
    'Contribution made to nonaffiliated committee',
    'In-kind contribution made to registered filer',
  ]
})
```

```python
contributions_df = contributions_df.join(transaction_df, how='left', on='transaction_tp')
```

```python
# source: https://www.fec.gov/campaign-finance-data/all-candidates-file-description/
candidate_col_names = ['cand_id', 'cand_name', 'cand_ici', 'pty_cd', 'cand_pty_affiliation', 'ttl_receipts', 'trans_from_auth', 'ttl_disb', 'trans_to_auth', 'coh_bop', 'coh_cop', 'cand_contrib', 'cand_loans', 'other_loans', 'cand_loan_repay', 'other_loan_repay', 'debts_owed_by', 'ttl_indiv_contrib', 'cand_office_st', 'cand_office_district', 'spec_election', 'prim_election', 'run_election', 'gen_election', 'gen_election_percent', 'other_pol_cmte_contrib', 'pol_pty_contrib', 'cvg_end_dt', 'indiv_refunds', 'cmte_refunds']

candidates_df = pl.read_csv('./data/weball26.txt', separator='|', has_header=False, new_columns=candidate_col_names, ignore_errors=True)
```

```python
# View dataframe
candidates_df
```

```python
candidates_df.null_count()
```

```python
# Join all dataframes
contributions_df_complete = contributions_df.join(pac_df, how='inner', on='cmte_id').join(candidates_df, how='inner', on='cand_id')
```

```python
deduped_interactions_df = contributions_df_complete[['cmte_id', 'cand_id', 'transaction_amt']].group_by(['cmte_id', 'cand_id']).sum()
```

```python
# Code to visualize transaction amount distribution
import matplotlib.pyplot as plt
import numpy as np

x = np.log(deduped_interactions_df.filter(pl.col('transaction_amt') > 1)['transaction_amt'].to_numpy())

plt.figure()
plt.hist(x, bins=50)
plt.xlabel("Log(Transaction Amount)")
plt.ylabel("Count")
plt.title("Distribution of Transaction Amounts")
plt.show()
```
